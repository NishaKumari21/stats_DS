```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom

# Bernoulli Distribution (p=0.5)
# Bernoulli distribution represents a single trial (like tossing a coin).
# p = 0.5 means there is a 50% chance of success (1) and 50% chance of failure (0).
# bernoulli.rvs(p, size=1000) generates 1000 random values (0s and 1s), simulating 1000 trials.
    
p = 0.5
bernoulli_data = bernoulli.rvs(p, size=1000)

# Binomial Distribution (n=10, p=0.5)
n = 10
binomial_data = binom.rvs(n, p, size=1000)

# Plotting
plt.figure(figsize=(12, 5))

# Bernoulli Plot
plt.subplot(1, 2, 1)
plt.hist(bernoulli_data, bins=2, rwidth=0.8, color='skyblue', edgecolor='black')
plt.xticks([0,1])
plt.title('Bernoulli Distribution')

# Binomial Plot
plt.subplot(1, 2, 2)
plt.hist(binomial_data, bins=n+1, rwidth=0.8, color='lightcoral', edgecolor='black')
# bins=n+1 ensures bins for all possible outcomes (0 to 10 heads in 10 tosses).
plt.xticks(range(n+1))
plt.title('Binomial Distribution')

plt.show()
# You will see two histograms:

# 1️⃣ Bernoulli Distribution (Left Graph) → Only two bars (for 0 and 1).
# 2️⃣ Binomial Distribution (Right Graph) → Multiple bars (for 0 to 10 heads).
```
